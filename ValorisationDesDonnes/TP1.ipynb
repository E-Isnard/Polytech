```r
#Data
data <- c(23,24,24,25,24,24,24,25,26,22,21)
#Center/Location
mean(data) # Calcule la moyenne
median(data) #Calcule la médiane
#Dispersion
sd(data) # Calcule l'écart-type
var(data)# Calcule la variance
IQR(data) # Calcule l'écart interquartile(Q3-Q1)
#Rank
min(data) # Calcule le minimum
max(data) # Calcule le maximum
quantile(data,1/4) # Calcule le 25% quantile
```

```r
n=30
k=1000
mu=5;sigma=2;sdclt=sigma/sqrt(n)
x = matrix(rnorm(n*k,mu,sigma),n,k) # Crée une matrice n*k valeurs d'une loi normale de paramètre (mu,sigma) 
#et les met dans une matrice
x.mean = apply(x,2,mean) # crée une vecteur des moyenne des colonnes
x.down = mu-4*sdclt;x.up=mu+4*sdclt;y.up=3/2
hist(x.mean,prob=T,xlim=c(x.down,x.up),ylim=c(0,y.up),main='Sampling distribution of the sample mean,Normal) case')

par(new=T)
x = seq(x.down,x.up,0.01)
y=dnorm(x,mu,sdclt)
plot(x,y,type='l',xlim=c(x.down,x.up),ylim=c(0,y.up))
```

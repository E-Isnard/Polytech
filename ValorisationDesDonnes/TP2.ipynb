```python
n=10
µ=0.3
s=1.4
x = rnorm(n,µ,s)
Fn = ecdf(x)
t = seq(-0.2,1.2,0.1)
plot(t,Fn(t),type="l",main="Comparaison entre Fn et F")
par(new=T)
F = pnorm(t,µ,s)
plot(t,F,type="l",col="blue",main="",xlab="",ylab="")
```

```python
eps=0.1
n=100
k = 1000
µ=0.3
s=1.4
t = seq(-0.2,1.2,0.1)
j = 0
for(i in 1:k)
{
    
    µ=0.3
    s=1.4
    x = rnorm(n,µ,s)
    Fn = ecdf(x)
    F = pnorm(t,µ,s)
    sup = max(abs(Fn(t)-F))
    # print(sup)
    if(sup>eps)
    {
        j=j+1
    }

}

print(c("Proba observée:",j/k))
print(c("Borne supérieure de DKW",2*exp(-2*n*eps*eps)))
```
